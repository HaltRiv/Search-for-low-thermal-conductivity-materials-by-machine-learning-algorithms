```python
import json
import pandas as pd
from urllib.request import urlopen
from ase.io import read
from dscribe.descriptors import MBTR
```

```python
API = 'http://aflow.org/API/aflux/'
MATCHBOOK = 'agl_thermal_conductivity_300K(*),volume_atom,volume_cell,spacegroup_relax,spacegroup_orig,geometry,positions_fractional,files'
DIRECTIVE = 'paging(0)'
REQUEST = API + '?' + MATCHBOOK + ',' + DIRECTIVE
print("aflux request: "+ REQUEST)
response = json.loads(urlopen(REQUEST).read().decode('utf-8'))
#print(json.dumps(response, indent = 4))
```

```python
data =pd.read_json(json.dumps(response, indent = 4))
data
```

```python
data_transposed = data.T
```

```python
data_transposed.to_csv("aflow_data_2",index = False)
```

```python
file = data_transposed.loc[data_transposed["compound"] =='Au4Ti1']['files']
```

```python
file[0][68]
```

```python
# Odczytaj plik CIF
atoms = read(file[0][68])
```

```python
# Zainicjuj deskryptor MBTR
mbtr = MBTR(
    species=[1, 6, 7, 8, 9],
    k1={
        "geometry": {"function": "atomic_number"},
        "grid": {"min": 0, "max": 1, "n": 20, "sigma": 0.1}
    },
    k2={
        "geometry": {"function": "inverse_distance"},
        "grid": {"min": 0, "max": 1/2, "n": 20, "sigma": 0.1},
        "weighting": {"function": "exp", "scale": 0.5, "cutoff": 1e-3}
    },
    k3={
        "geometry": {"function": "cosine"},
        "grid": {"min": -1, "max": 1, "n": 10, "sigma": 0.1},
        "weighting": {"function": "exp", "scale": 0.5, "cutoff": 1e-3}
    },
    periodic=True,
    normalization='l2_each'
)

# Oblicz deskryptory dla układu atomów
mbtr_descriptors = mbtr.create(atoms)
```
