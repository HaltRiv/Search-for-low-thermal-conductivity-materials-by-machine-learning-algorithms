```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
krr_w_MN = pd.read_csv('KRR wihout MN.csv')
krr_w_MN

tcets_w_MN = [['atomic number','spacegroup relax'],
              ['group'],
              ['volume_cell'],
              ['thermal conductivity'],
              ['valence electrons','electron affinity'],
              ["absolute radii"],
              ['molar volume','the orbital exponent of Slater-type orbitals'],
              ['polarizability'],
              ['electronegativity'],
              ['volume_atom',],
              ['boiling point'],
              ['fusion enthalpy','covalent radii'],
              ['atomization enthalpy','global hardness','atomic density'],
              ['melting point','van der Waals radii','first ionization energy','electrophilicity indices']

]
```

```python

def ploter_R2(df,Title):
    #fig, ax = plt.subplots(layout='constrained')
    plt.figure(figsize=(20, 10))
    plt.plot(df['corelation'], df['R_Sq_train'], label='Train',marker='o')
    plt.plot(df['corelation'], df['R_Sq_test'], label='Test',marker='^')
    plt.title(Title)
    plt.xlabel('Corelation')
    plt.ylabel('R^2')
    #plt.legend(label = ["Train",'test'])
    plt.legend()
    plt.xticks(df['corelation'], ['[-]',"[-]","[-]","[-]","[-]","[-]","[-]","[-]","[-]","[-]","thermal conductivity\n"
                                                                                              "electron affinity","[-]","thermal conductivity","volume cell","'group'","[-]","[-]","atomic number,\nspacegroup relax",])
    plt.show()
def ploter_MSE(df,Title):
    plt.show()
    df.plot(x = 'corelation',y = ['MSE_train','MSE_test'],figsize=(10,5))
    plt.title(Title)
    plt.xlabel('Corelation')
    plt.ylabel('Mean Square Error')
    #plt.legend(label = ["Train",'test'])
    #plt.legend(labels ='Test')
    plt.show()
```

```python
ploter_R2(krr_w_MN,"Kernel ridge regression without Mendelev Number")
```

```python
ploter_MSE(krr_w_MN,"Kernel ridge regression without Mendelev Number")
```
